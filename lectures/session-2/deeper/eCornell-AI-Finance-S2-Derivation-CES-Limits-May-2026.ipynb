# Derivation: Limits of the CES Utility Family

In this derivation, we establish the three textbook limits of the Constant Elasticity of Substitution (CES) utility family used by the lecture. We derive the closed-form CES optimum in share counts, then take $\eta \to 1$, $\eta \to \infty$, and $\eta \to 0^{+}$ to recover Cobb-Douglas, linear, and Leontief allocators, respectively.

> __Learning Objectives:__
>
> By the end of this notebook, we will be able to:
> * __CES first-order condition:__ We will derive the closed-form CES share-count optimum from the budget-constrained Lagrangian. We will see how the elasticity $\eta$ enters as an exponent on the bang-for-the-buck ratio.
> * __Cobb-Douglas as the unit-elasticity limit:__ We will show that the unit-elasticity case collapses CES to Cobb-Douglas using a standard L'Hôpital argument. We will read off the dollar-weight optimum that is proportional to the preference weights as a special case of the CES formula.
> * __Linear and Leontief endpoints:__ We will show that the high-elasticity limit concentrates the book on the single best asset and that the low-elasticity limit equalizes share counts. We will be able to predict the qualitative shape of the optimum from $\eta$ alone.

___

## The CES Utility and Budget Constraint

Recall the CES utility function over share counts $\mathbf{n} = (n_{1}, \ldots, n_{K})$ for the preferred set $\mathcal{A}^{+} = \{i : \gamma_{i} > 0\}$:
$$
U_{\text{CES}}(\mathbf{n}) = \left(\sum_{i \in \mathcal{A}^{+}} \gamma_{i}\, n_{i}^{\rho}\right)^{1/\rho}, \qquad \rho = \frac{\eta - 1}{\eta}
$$
where $\eta > 0$ is the elasticity of substitution and the preference weights satisfy $\gamma_{i} \in (0, 1)$.

The allocator maximizes $U_{\text{CES}}$ subject to the budget constraint:
$$
\sum_{i \in \mathcal{A}^{+}} P_{i}\, n_{i} = B
$$
where $P_{i}$ is the price of asset $i$ and $B$ is the budget. The Lagrangian for this constrained maximization is given by:
$$
\mathcal{L}(\mathbf{n}, \mu) = \left(\sum_{i \in \mathcal{A}^{+}} \gamma_{i}\, n_{i}^{\rho}\right)^{1/\rho} - \mu\left(\sum_{i \in \mathcal{A}^{+}} P_{i}\, n_{i} - B\right)
$$
where $\mu$ is the Lagrange multiplier on the budget constraint.

___

## First-Order Conditions

Let $S = \sum_{j \in \mathcal{A}^{+}} \gamma_{j}\, n_{j}^{\rho}$ so that $U = S^{1/\rho}$. Differentiating $U$ with respect to $n_{i}$ via the chain rule yields:
$$
\frac{\partial U}{\partial n_{i}} = \frac{1}{\rho}\, S^{1/\rho - 1} \cdot \rho\, \gamma_{i}\, n_{i}^{\rho - 1} = S^{1/\rho - 1}\, \gamma_{i}\, n_{i}^{\rho - 1}
$$

Setting $\partial \mathcal{L} / \partial n_{i} = 0$ gives the FOC for each preferred asset $i \in \mathcal{A}^{+}$:
$$
S^{1/\rho - 1}\, \gamma_{i}\, n_{i}^{\rho - 1} = \mu\, P_{i}
$$

Taking the ratio of the FOC for assets $i$ and $j$ cancels the common factor $S^{1/\rho - 1}$:
$$
\frac{\gamma_{i}\, n_{i}^{\rho - 1}}{\gamma_{j}\, n_{j}^{\rho - 1}} = \frac{P_{i}}{P_{j}}
$$

Rearranging and using $\rho - 1 = -1/\eta$ gives:
$$
n_{i}^{-1/\eta} \propto \frac{P_{i}}{\gamma_{i}} \quad \Longleftrightarrow \quad n_{i} \propto \left(\frac{\gamma_{i}}{P_{i}}\right)^{\eta}
$$

Imposing the budget constraint $\sum_{i} P_{i}\, n_{i} = B$ and solving for the proportionality constant yields the closed-form CES share-count optimum:
$$
\boxed{\;n_{i}^{\star} = \frac{(\gamma_{i} / P_{i})^{\eta}}{\sum_{j \in \mathcal{A}^{+}} P_{j}\, (\gamma_{j} / P_{j})^{\eta}}\, B\quad\blacksquare\;}
$$

The corresponding dollar weights $w_{i}^{\star} = P_{i}\, n_{i}^{\star} / B$ have the standard CES expenditure-share form:
$$
\boxed{\;w_{i}^{\star} = \frac{P_{i}^{1-\eta}\, \gamma_{i}^{\eta}}{\sum_{j \in \mathcal{A}^{+}} P_{j}^{1-\eta}\, \gamma_{j}^{\eta}}\;}
$$

> __Reading the formula:__
>
> The exponent $\eta$ on $\gamma_{i} / P_{i}$ is the lever the lecture's three limits work on. Larger $\eta$ amplifies the spread between bang-for-the-buck ratios; smaller $\eta$ flattens it. The limits below take $\eta$ to its three extremes and read off the resulting allocator shape.

___

## Limit 1: $\eta \to 1$ recovers Cobb-Douglas

We first verify that the CES utility itself converges to Cobb-Douglas. For $\eta \to 1$ we have $\rho \to 0$, and the limit
$$
\lim_{\rho \to 0}\, \left(\sum_{i \in \mathcal{A}^{+}} \gamma_{i}\, n_{i}^{\rho}\right)^{1/\rho}
$$
is a $1^{\infty}$ indeterminate form. Taking the logarithm gives:
$$
\log U_{\text{CES}} = \frac{1}{\rho}\, \log\!\left(\sum_{i \in \mathcal{A}^{+}} \gamma_{i}\, n_{i}^{\rho}\right)
$$

Both numerator and denominator vanish as $\rho \to 0$ when the preference weights are normalized so $\sum_{i} \gamma_{i} = 1$. Applying L'Hôpital's rule with respect to $\rho$ yields:
$$
\lim_{\rho \to 0} \log U_{\text{CES}} = \frac{\sum_{i \in \mathcal{A}^{+}} \gamma_{i}\, n_{i}^{\rho}\, \log n_{i}}{\sum_{i \in \mathcal{A}^{+}} \gamma_{i}\, n_{i}^{\rho}}\bigg|_{\rho=0} = \sum_{i \in \mathcal{A}^{+}} \gamma_{i}\, \log n_{i}
$$

Exponentiating recovers the Cobb-Douglas form:
$$
\boxed{\;\lim_{\eta \to 1} U_{\text{CES}}(\mathbf{n}) = \prod_{i \in \mathcal{A}^{+}} n_{i}^{\gamma_{i}}\quad\blacksquare\;}
$$

The CES optimum collapses correspondingly. Setting $\eta = 1$ in the dollar-weight formula gives:
$$
w_{i}^{\star} = \frac{P_{i}^{1-1}\, \gamma_{i}^{1}}{\sum_{j} P_{j}^{1-1}\, \gamma_{j}^{1}} = \frac{\gamma_{i}}{\sum_{j \in \mathcal{A}^{+}} \gamma_{j}}
$$

so dollar weights track preference weights and are independent of prices. The marginal-utility-per-dollar interpretation: at the optimum, $\gamma_{i} / (P_{i}\, n_{i}^{\star})$ is equal across all preferred assets, so each asset delivers the same marginal utility per dollar of budget.

___

## Limit 2: $\eta \to \infty$ concentrates on the best asset

For $\eta \to \infty$ we have $\rho \to 1$, and the CES utility itself becomes linear:
$$
\boxed{\;\lim_{\eta \to \infty} U_{\text{CES}}(\mathbf{n}) = \sum_{i \in \mathcal{A}^{+}} \gamma_{i}\, n_{i}\;}
$$

To see the optimum, fix the closed-form CES share counts and let $\eta \to \infty$. Define the maximum bang-for-the-buck ratio $r^{\star} = \max_{i \in \mathcal{A}^{+}} \gamma_{i} / P_{i}$ and let $i^{\star} = \arg\max_{i} \gamma_{i} / P_{i}$. Factoring $r^{\star}$ out of the numerator and denominator of $n_{i}^{\star}$ gives:
$$
n_{i}^{\star} = \frac{(\gamma_{i} / P_{i})^{\eta}}{\sum_{j} P_{j}\, (\gamma_{j} / P_{j})^{\eta}}\, B = \frac{(r_{i} / r^{\star})^{\eta}}{\sum_{j} P_{j}\, (r_{j} / r^{\star})^{\eta}}\, B
$$
where $r_{i} = \gamma_{i} / P_{i}$.

For $i \neq i^{\star}$, the ratio $r_{i} / r^{\star} < 1$, so $(r_{i} / r^{\star})^{\eta} \to 0$ as $\eta \to \infty$. Only the $i = i^{\star}$ term survives in the denominator, where $(r_{i^{\star}} / r^{\star})^{\eta} = 1$. The optimum therefore concentrates on the single best asset:
$$
\boxed{\;\lim_{\eta \to \infty} n_{i}^{\star} = \begin{cases} B / P_{i^{\star}} & i = i^{\star} \\ 0 & \text{otherwise} \end{cases}\quad\blacksquare\;}
$$

This is the standard linear-programming corner solution: when utility is linear in share counts, the budget-constrained maximum spends every dollar on whichever asset has the highest preference weight per dollar of price.

___

## Limit 3: $\eta \to 0^{+}$ equalizes share counts

For $\eta \to 0^{+}$ we have $\rho \to -\infty$. With $\rho < 0$, the term $n_{i}^{\rho} = 1 / n_{i}^{|\rho|}$ is largest for the smallest $n_{i}$, so the sum $\sum_{j} \gamma_{j}\, n_{j}^{\rho}$ is dominated by the term with the smallest share count. Taking the $1/\rho$ root with $\rho \to -\infty$ extracts the minimum:
$$
\boxed{\;\lim_{\eta \to 0^{+}} U_{\text{CES}}(\mathbf{n}) = \min_{i \in \mathcal{A}^{+}} n_{i}\;}
$$
up to a constant factor that is absorbed into the optimization. This is the Leontief, or perfect-complements, form.

The optimum follows directly from the closed-form CES share counts. Setting $\eta = 0$ in the share-count formula yields $(\gamma_{i} / P_{i})^{0} = 1$ for all preferred assets, so:
$$
\boxed{\;\lim_{\eta \to 0^{+}} n_{i}^{\star} = \frac{B}{\sum_{j \in \mathcal{A}^{+}} P_{j}}\quad\blacksquare\;}
$$
for every $i \in \mathcal{A}^{+}$. All preferred assets receive the same share count, independent of $\gamma_{i}$ or $P_{i}$.

In dollar-weight terms, $w_{i}^{\star} = P_{i} / \sum_{j} P_{j}$, which is price-weighted rather than equal-dollar. Equal allocation here means equal share counts, not equal dollar exposure: the Leontief allocator treats preferred assets as perfect complements that the optimum wants in 1:1 share-count ratio.

___

## Summary

The CES utility family has a single closed-form share-count optimum, $n_{i}^{\star} \propto (\gamma_{i} / P_{i})^{\eta}$, that interpolates between three textbook allocators as $\eta$ moves across its range. We derived the FOC from the budget-constrained Lagrangian, then took the three limits to recover Cobb-Douglas at $\eta = 1$, the linear corner solution at $\eta \to \infty$, and the equal-share-count Leontief allocator at $\eta \to 0^{+}$.

> __Why this matters:__
>
> The lecture treats $\eta$ as a single scalar lever that reshapes the engine's preferences without re-solving the underlying program. The three limits show that this lever spans a meaningful axis: from price-weighted equal shares at one extreme, through the dollar-weighted Cobb-Douglas optimum in the middle, to all-in concentration at the other extreme. Session 3's eta-bandit is searching this axis.

> __Key Takeaways:__
>
> * __One formula, three regimes:__ The closed-form CES optimum $n_{i}^{\star} \propto (\gamma_{i} / P_{i})^{\eta}$ is a single expression whose qualitative shape is set entirely by $\eta$. Reading $\eta$ alone tells us whether the allocator concentrates, equalizes, or splits the budget by preference weights.
> * __Cobb-Douglas is the unit-elasticity center:__ At $\eta = 1$ the CES utility collapses to a Cobb-Douglas product and dollar weights track preference weights. The L'Hôpital limit confirms that this is a special case of CES, not a separate model.
> * __Endpoints are corner solutions:__ At $\eta \to \infty$ the optimum spends every dollar on $\arg\max_{i} \gamma_{i} / P_{i}$; at $\eta \to 0^{+}$ it equalizes share counts regardless of preferences. These two extremes bound the entire CES family.
